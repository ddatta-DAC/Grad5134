```python
import re
import csv
```

```python
checkin2commu = {}
fl = open('chicago_checkins_data.csv', 'r')
reader = csv.reader(fl, delimiter=',')
for line in reader:
    print line
    break
for line in reader:
    checkin2commu[line[0]] = line[6]
fl.close()
```

```python
cmm_usr = {str(k): {} for k in range(1, 78)}
cmm_loc = {str(k): {} for k in range(1, 78)}
cmm_chn = {str(k): 0 for k in range(1, 78)}
fl = open('ratings.csv', 'r')
reader = csv.reader(fl, delimiter=',')
for line in reader:
    print line
    break
for line in reader:
    try:
        cmm_usr[checkin2commu[line[2]]][line[0]] += 1
    except:
        cmm_usr[checkin2commu[line[2]]][line[0]] = 1
    try:
        cmm_loc[checkin2commu[line[2]]][line[1]] += 1
    except:
        cmm_loc[checkin2commu[line[2]]][line[1]] = 1
    cmm_chn[checkin2commu[line[2]]] += 1
fl.close()
print cmm_usr
print cmm_loc
print cmm_chn
```

```python
fusr = open('commu_usrs.csv', 'w')
fusr.write('commu_area,count\n')
for itm in cmm_usr:
    fusr.write(','.join([itm, str(len(cmm_usr[itm]))])+'\n')
fusr.close()

fusr = open('commu_locations.csv', 'w')
fusr.write('commu_area,count\n')
for itm in cmm_loc:
    fusr.write(','.join([itm, str(len(cmm_loc[itm]))])+'\n')
fusr.close()

fusr = open('commu_checkin.csv', 'w')
fusr.write('commu_area,count\n')
for itm in cmm_chn:
    fusr.write(','.join([itm, str(cmm_chn[itm])])+'\n')
fusr.close()
```
